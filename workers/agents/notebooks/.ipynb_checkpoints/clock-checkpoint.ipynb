```python
from pathlib import Path

data_dir = Path("./results-06-09")
list(data_dir.iterdir())
```

```python
import pandas as pd
import json

experiments = {}

for dir in data_dir.iterdir():
    print(dir)
    settings = json.loads((dir / "settings.json").read_text())
    key = (settings["workers"], settings["stale"])
    print(key)
    experiments[key] = [pd.read_csv(csv_file) for csv_file in dir.iterdir() if csv_file.name.endswith(".csv")]
```

```python
experiments
```

```python
for experiment, data in experiments.items():
    initial_time = min(worker["time_refs"].iloc[0] for worker in data)
    final_time = min(worker["time_refs"].iloc[-2] for worker in data)
    
    time_refs = [worker["time_refs"] for worker in data]
    print(experiment)
    for i in range(int(initial_time) + 1, int(final_time) + 1, 200):
        try:
            clocks = [ref[ref < i].idxmax() for ref in time_refs]
            print(clocks)
        except ValueError:
            continue
```
