```python
# exp: [24/09, 17/10]
aggregated_jct = {
    (4, 0): [95080.3899345398, 97548.66936039925],
    (4, 10): [57346.6244866848, 56323.38115429878],
    (4, 16): [57132.38931059837, 56980.16577172279],
    (4, 32): [55815.76748275757, 54805.50754547119],
    (4, 48): [55513.89518260956, 55037.356454372406],
    (4, 500): [52804.435685396194, 54893.40288162232],
}
```

```python
from statistics import mean

avg_jct = {}
for exp, jcts in aggregated_jct.items():
    avg_jct[exp] = mean(jcts)

max_jct = max(avg_jct.values())
normalized_jct = {exp: time * 100 / max_jct for exp, time in avg_jct.items()}
plot_time(normalized_jct, "JCT(%)", "average_jct.pdf")
```

```python
# exp: [24/09, 17/10]
aggregated_wt = {
    (4, 0): [66951.84495973587, 69321.2718129158],
    (4, 10): [34614.42370700836, 33736.30786252022],
    (4, 16): [34561.99371385574, 34248.32032704353],
    (4, 32): [33161.17557120323, 33325.541080236435],
    (4, 48): [33382.95794892311, 32778.52022957802],
    (4, 500): [32079.161390781403, 32957.41082692146],
}
```

```python
avg_wt = {}
for exp, wts in aggregated_wt.items():
    avg_wt[exp] = mean(wts)

max_wt = max(avg_wt.values())
normalized_wt = {exp: time * 100 / max_jct for exp, time in avg_wt.items()}
plot_time(normalized_wt, "WT(%)", "average_wt.pdf", ylim=[0, 105])
```

```python
processing_time = {exp: jct - wt for (exp, jct), (_, wt) in zip(normalized_jct.items(), normalized_wt.items())}
plot_time(processing_time, "TT(%)", "processing_time.pdf", ylim=[0, 105])
```
